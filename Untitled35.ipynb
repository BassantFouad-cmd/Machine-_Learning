```python
#### task 2.1
 
class Library:
    def __init__(self, name, books_list):
        self.name = name
        self.books_list = books_list
        self.lended_books = {}
 
    def lend_a_book(self, user_name, book_name):
        if book_name in self.books_list:
            if user_name in self.lended_books:
                self.lended_books[user_name].append(book_name)
            else:
                self.lended_books[user_name] = [book_name]
            self.books_list.remove(book_name)
            print(f"{book_name} has been lent to {user_name}.")
        else:
            # Check if the book is already lent
            for user, books in self.lended_books.items():
                if book_name in books:
                    print(f"Sorry, {book_name} is already lent to {user}.")
                    return
            print(f"Sorry, {book_name} is not available.")
 
    def return_borrowed_book(self, user_name, book_name):
        if user_name in self.lended_books and book_name in self.lended_books[user_name]:
            self.lended_books[user_name].remove(book_name)
            if not self.lended_books[user_name]:
                del self.lended_books[user_name]
            self.books_list.append(book_name)
            print(f"Thanks!, for delivering the book {book_name}.")
        else:
            print(f"Sorry, {book_name} was not lent to {user_name}.")
 
    def display_available_books(self):
        print("Available books:")
        for book in self.books_list:
            print(book)
 
    def display_lended_books(self):
        print("Lended books:")
        for user, books in self.lended_books.items():
            print(f"{user}: {books}")
 
    def add_book_to_library(self, book_name):
        self.books_list.append(book_name)
        print(f"{book_name} has been added to the library.")
 
# Create an object harry with the specified attributes
harry = Library("CodeWithHarry", ["Python", "Rich Daddy Poor Daddy", "Harry Potter", "C++ Basics", "Algorithms by CLRS"])
 
# Display initial available books
print("Initial available books:")
harry.display_available_books()
print("\n")
 
# Lend books
harry.lend_a_book('ahmed', 'Python')
harry.lend_a_book('mohamed', 'C++ Basics')
harry.lend_a_book('said', 'Rich Daddy Poor Daddy')
harry.lend_a_book('ahmed', 'Harry Potter')
 
# Try to lend a book that is already lent
harry.lend_a_book('mohamed', 'Python')
 
# Display lended books
print("\nLended books:")
harry.display_lended_books()
 
# Display available books again
print("\nAvailable books after lending:")
harry.display_available_books()
 
# Return a borrowed book
harry.return_borrowed_book('ahmed', 'Python')
 
# Display lended books again
print("\nLended books after returning:")
harry.display_lended_books()
 
# Display available books again
print("\nAvailable books after returning:")
harry.display_available_books()
 
# Add a new book to the library
harry.add_book_to_library("Animals")
 
# Display available books after adding a new book
print("\nAvailable books after adding a new book:")
harry.display_available_books()
```
