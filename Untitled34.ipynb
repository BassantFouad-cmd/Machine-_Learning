```python


class Library:

    def __init__(self, library_name, booklist):
        self.library_name = library_name
        self.booklist = booklist
        self.__lent_books_dictionary = { }

    def lend_books(self, book_name, duration, customer_name):

        if book_name in self.__lent_books_dictionary.items():
            print(f"The book '{book_name}' is already lent to {self.__lent_books_dictionary[book_name]}.")
        elif book_name in self.booklist:
            self.__lent_books_dictionary[book_name] = customer_name
            print(f"The book '{book_name}' has been successfully lent to {customer_name} for {duration} days.")
        else:
            print(f"The book '{book_name}' is not available in the library.")


    def get_lent_books_dictionary(self):
        self.get_lent_books_dictionary = self.__lent_books_dictionary
        print(f"the lent books dictionary is {self.__lent_books_dictionary}")
    
    def available_books(self):
        self.booklist = set(self.booklist)
        available_books = self.booklist.difference(set(self.__lent_books_dictionary.keys()))
        print (f"available books are : {available_books}")
        
    
    def return_books(self,returned_books,customer_name):
        self.returned_books = returned_books
        try:
            if self.returned_books in self.__lent_books_dictionary.keys() and self.__lent_books_dictionary[returned_books].lower()== customer_name.lower():
                del self.__lent_books_dictionary[returned_books]
                print(f"the {self.returned_books} has been returned successfully ,thanks!, for delivering the book")
            else:
                print (f"The book '{returned_books}' was not borrowed by {customer_name} or is not in the lent books list.")
        except KeyError:
            print("Please write a valid Customer name")


    def add_book(self,book_name):
        if book_name in self.booklist:
            print(f"The book '{book_name}' is already in the library.")
        else:
            self.booklist = list (self.booklist)
            self.booklist.append(book_name)
            print(f"The book '{book_name}' has been successfully added to the library and here is the book list {self.booklist}")



def library_menu():
    library_name = input("Enter the name of the library: ")
    list_books = input("Enter the initial book list (comma-separated): ").split(",")
    list_books = [book.strip() for book in list_books]  
    library = Library(library_name, list_books)
    
    
    while True:
        print ("Welcome to our Library, please type a number from the next options:\n1) Display Books\n2) Lend a Book\n3) Add a Book\n4) Return a Book\n5) To Exit\n")
        User_Choice = input("Please write your Choice number")
            
        if User_Choice == "1" :
            library.available_books()
            print(150*"_")

        elif User_Choice == "2" :
            book_name = input("Enter the book name you want to lend: ")
            customer_name = input("Enter your name: ")
            duration = input("Enter the duration (in days): ")
            library.lend_books(book_name, duration, customer_name)
            print(150*"_")
                
        elif User_Choice == "3" :
            book_name = input("Enter the book name you want to add: ")
            library.add_book(book_name)
            print(150*"_")


        elif User_Choice == "4" :
            returned_books = input("Enter the book name you want to return: ")
            customer_name = input("Enter your name: ")
            library.return_books(returned_books, customer_name)
            print(150*"_")

        elif User_Choice == "5" :
            print("Goodbye, have a nice day !!")
            print(150*"_")
            break

        else:
            print("Invalid choice. Please try again.")
            print(150*"_")

library_menu()
```
