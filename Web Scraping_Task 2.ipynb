```python
from bs4 import BeautifulSoup
import requests
response = requests.get("https://forecast.weather.gov/MapClick.php?lat=37.7772&lon=-122.4168")
response.text
```

```python
soup = BeautifulSoup(response.text, 'html.parser')

forecast_elements = soup.find_all('div', class_='tombstone-container')


for element in forecast_elements:
    period_name = element.find('p', class_='period-name')
    temp_high = element.find('p', class_='temp temp-high')
    temp_low = element.find('p', class_='temp temp-low')
    
    period_name_text = period_name.text if period_name else "N/A"
    temp_high_text = temp_high.text if temp_high else "N/A"
    temp_low_text = temp_low.text if temp_low else "N/A"
    
    print(f"Period: {period_name_text}")
    print(f"High Temperature: {temp_high_text}")
    print(f"Low Temperature: {temp_low_text}")
    print("-" * 40)
```
